```python
# cheking null data 
df.info()
```

```python
# cheking duplicate data
df.drop_duplicates()
```

```python
#grouping by stores
df1=df.groupby('store').sum('customer_visits')
df1.head()
```

```python
def_divice=df1[df1['customer_visits']==0]
```

```python
# stores with broken or off visit counter machines.
def_divice.head()
```

```python
def_divice=def_divice.rename(columns={"receipts": "receipts_sum", "customer_visits": "customer_visits_sum"})
```

```python
def_divice.head()
```

```python
defected_div=df.merge(def_divice, on='store')
```

```python
defected_div.head()
```

```python
defected_div.to_csv('defected_devices.csv', index=False)
```

```python
df.head()
```

```python
ndef_divice=df1[df1['customer_visits']!=0]
ndef_divice.head()
```

```python
# stores with healthy visit counter machines
ndef_divice=ndef_divice.rename(columns={"receipts": "receipts_sum", "customer_visits": "customer_visits_sum"})
```

```python
ndef_divice.head()
```

```python
ndefected_div=df.merge(ndef_divice, on='store')
ndefected_div.head()
```

```python
ndefected_div.to_csv('Non_defected_devices.csv', index=False)
```
