# Competitor_marketing_spend

## Let's check the impact of Competitor_marketing_spend

```python
df=df.sort_values(by=['competitor_marketing_spend'], ascending=False)
```

```python
#scatter plot with budget vs gross
plt.scatter(x=df['customer_visits'],y=df['competitor_marketing_spend'])
plt.title('competitor_marketing_spend vs Customer visit')
plt.xlabel('Customer visit')
plt.ylabel('competitor_marketing_spend')
plt.show()
```

```python
#plot the precipation durtaion vs customer visit
sns.regplot(x=df['customer_visits'],y=df['competitor_marketing_spend'],data=df,scatter_kws={"color":"red"},
            line_kws={"color":"blue"})
```

## Result: A weak positive correlation between customer_visits and competitor_marketing_Spend

```python
correlation_matrix=df.corr()
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.show()
```

### This matrix shows that "Customer visit" has weak positive correlation with "competitor_marketing Spend" and "folder_pages" and very weak negetive correlation with "online_ad_impressions,radio_ads_reach and precipation_duration"
